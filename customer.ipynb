there are null values in the column MINIMUM_PAYMENTS ,we should fill with string '0'                   

```python
data = data.fillna('0')
```

```python
data = data.drop('CUST_ID', axis=1)
```

```python
data.head()
```

Correlation matrix

```python
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f",mask=np.triu(data.corr()) )
plt.title('Correlation Matrix')
plt.show()
```

* higher the correlation factor higher the high chance to droping one of two columns 
- we can see PURCHASES and ONEOFF_PURCHASES are highly correlated,  we can drop ONEOFF_PURCHASES to avoid curse of dimensionality.

```python
data = data.drop('ONEOFF_PURCHASES', axis= 1)
```

```python
data.shape
```

```python
data.info()
```

changing datatype 'object' into 'float' 

```python
data = data.astype(float)
```

```python
data.info()
```

### Ploting subplots 

```python
# Visualize the distribution of numerical variables
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 16))
fig.suptitle('Distribution of Numerical Variables')

for i, col in enumerate(data.columns):
    try:
        sns.histplot(data[col].dropna(), ax=axes[i // 4, i % 4], kde=True)
        axes[i // 4, i % 4].set_title(col)
    except Exception as e:
        print(f"Error plotting {col}: {e}")

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to avoid overlap
plt.show()
```

data is right skewed so we are appling log transformation to make it normally distrubuted

```python
for i, col in enumerate(data.columns):
    data[col] = np.log1p(data[col])
```

we again visualising to verify

```python
# Visualize the distribution of numerical variables
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 16))
fig.suptitle('Distribution of Numerical Variables') 

for i, col in enumerate(data.columns):
    try:
        sns.histplot(data[col].dropna(), ax=axes[i // 4, i % 4], kde=True) 
        axes[i // 4, i % 4].set_title(col)
    except Exception as e:
        print(f"Error plotting {col}: {e}")

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to avoid overlap
plt.show() 
```
