```python
data = data.drop('CUST_ID', axis=1)
```

```python
data.head()
```

```python
# Visualize the distribution of numerical variables
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 16))
fig.suptitle('Distribution of Numerical Variables')

for i, col in enumerate(data.select_dtypes(include=['float64', 'int64']).columns):
    try:
        sns.histplot(data[col].dropna(), ax=axes[i // 4, i % 4], kde=True)
        axes[i // 4, i % 4].set_title(col)
    except Exception as e:
        print(f"Error plotting {col}: {e}")

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to avoid overlap
plt.show()
```

```python
# Correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f",mask=np.triu(data.corr()) )
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualize the original distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['CREDIT_LIMIT'].dropna(), kde=True, color='blue')
plt.title('Original Distribution')
plt.show()

# Apply log transformation
log_transformed_column = np.log1p(data['CREDIT_LIMIT'])

# Visualize the transformed distribution
plt.figure(figsize=(10, 6))
sns.histplot(log_transformed_column.dropna(), kde=True, color='green')
plt.title('Log Transformed Distribution')
plt.show()
```
